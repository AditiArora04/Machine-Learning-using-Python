```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv")
df.head()
```

```python
df.shape
```

```python
df1=df.drop("date",axis=1)
df1.head()
```

```python
#seperate daaset into independet and dependent variables
```

```python
x=df1.iloc[:,:4].values
x
```

```python
y=df1.iloc[:,4].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
x_test
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred #predicted value
```

```python
y_test 
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
#sum of right diagonal show how many sample matches (here 121 sample matched)
```

```python
y_test.size 
#out of 147 , 121 are matched (82% accuracy rate)
```
