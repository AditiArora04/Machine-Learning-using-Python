```python
import matplotlib.pyplot as plt 
from matplotlib import style
import pandas as pd
style.use('ggplot')
#ggplot- style scheme 
```

```python
#Example-1 
x=[1,2,3]
y=[4,5,1]
plt.plot(x,y,'y')
plt.title("Info")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
plt.plot([1,2,3],[4,5,1],'b')
plt.show()
```

```python
# Example -2
x1=[5,8,10]
y1=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x1,y1,'g',label="Line One",linewidth=5)
plt.plot(x2,y2,'r',label="Line One",linewidth=5)
plt.legend() # to show label
plt.title('Epic Info')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True,color='k')
plt.show()
```

```python
# Creation of Bar Graph
days1=[0.25,1.25,2.25,3.25,4.25] # Days for BMW
km1=[50,40,70,80,20] # Km. for BMW
days2=[0.75,1.75,2.75,3.75,4.75] # Days for Audi
km2=[80,20,20,50,60] # Km. for Audi
plt.bar(days1,km1,color='y',label='BMW',width=0.4)
plt.bar(days2,km2,color='r',label='Audi',width=0.4)
plt.legend()
plt.title('Information')
plt.xlabel('Days')
plt.ylabel('Distance(Kms)')
plt.show()
```
